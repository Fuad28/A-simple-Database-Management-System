## A SIMPLE DATABASE MANAGEMENT SYSTEM

1. MODE OF USERS:
        a. Admin
        b. Student
        
2. FUNCTIONALITIES:
        a. Admin can:
            i. Store students/ sign_up
            ii. Update students info
            iii. Delete students records
            iv. Retrieve students Data
            v. View records (can see all students ID but cannot see other admins ID)
        b. Students can:
            i. Store their records/ sign_up
            ii. Update their info
            iii. Retrieve their Data
            iv. View records (can see their ID)
            
            
 3. TRICK:
 
       a. You can quit/ exit the programme 
            i. by entering 'stop' or nothing whenever you're prompted to input something
            ii. By entering 'No' at points where you're asked.
            
       b. You start the script by running 'architecure_tier_1()'
         
       c. After registering, you need to login again with your first name, last names and new ID before you can perform any actions

```python
class DSC_database:
    import pandas as pd
    stud_id, admin_id= ['ST_'+str(i) for i in list(range(1000, 20001))], ['AD_'+str(i) for i in list(range(100, 151))]
    database= pd.DataFrame(columns= ['Name', 'Email', 'Id', 'Mode']).set_index('Id')

    def __init__(self, first, last, email, i_d= None):
         self.first, self.last, self.email, self.i_d=  first, last, email, None
    
    def Fullname(self):
        return self.first +' '+ self.last
    
    def Email(self):
        return self.email
        
    def sign_up(self, mode):
        import random
        if mode=='student':
            i_d= random.choice(self.stud_id)
            self.i_d= i_d
            self.stud_id.remove(i_d)
            self.database.loc[self.i_d]= {'Name': self.Fullname(), 'Email': self.Email(), 'Mode': mode}
            print(f'\nWelcome {self.Fullname()}!!!\nYour ID is {i_d}')
            return i_d
        elif mode=='admin':
            i_d= random.choice(self.admin_id)
            self.i_d= i_d
            self.admin_id.remove(i_d)
            self.database.loc[self.i_d]= {'Name': self.Fullname(), 'Email': self.Email(), 'Mode': mode}
            print(f'\nWelcome!!!\nYour ID is {i_d}')
            return i_d
    
    def retrive_data(self, i_d, mode):
        if mode== 'admin':
            if i_d in self.database.index:
                return self.database.loc[i_d]
            else:
                return '\nUNREGISTERED ID'
        elif mode== 'student':
            if i_d in self.database.index and (i_d.startswith('a') == False):
                return self.database.loc[i_d]
            else:
                print('\nUNREGISTERED ID')
                   
    def delete_data(self, i_d):
        if i_d in self.database.index:
            self.database.drop(i_d, 0, inplace= True)
            print('\nDONE!!!')
        else:
            print('\nUNREGISTERED ID')
    
    def update_data(self, i_d, name, email):
        self.name, self.email= name, email
        if i_d in self.database.index:
            self.database.loc[i_d].Name= self.name
            self.database.loc[i_d].Email= self.email
            print(f'\nIndividual with ID {i_d} updated')
        else:
            print('UNREGISTERED ID')
    
    def view_data(self, mode, i_d):
        import pandas as pd
        if mode=='admin' and i_d in self.database.index:
          return print(self.database)
        elif mode== 'student' and i_d in self.database.index:
            return print(self.database[self.database.Mode=='student'])
        else:
            print('Either unidentified mode or i_d, Try again!!')
```

Now we're done with the class, let's go on to build the architecture

```python
clearance= '''You've the clearance to:
\t1. Store students/ sign_up
\t2. Update students info
\t3. Retrieve students Data
\t4. Delete data (strictly for admins)
\t5. View data

Reply with 1 -5 for your desired action'''
    
    
def stop_tier_1():
    import sys
    ask= input('\nDo you wish to continue: ')
    if ask.lower()=='yes':
        return architecure_tier_1()
    else:
        print('okay then, bye')
        return sys.exit()
    

def stop_tier_2(item):
    if (item== '' or item.lower()== 'stop'):
        print('\nOOps!!! Something has gone wrong')
        stop_tier_1()
    else:
        pass

    
def architecure_tier_1():
    print('\n\n========================Welcome to MY DATABASE MANAGEMENT SYSTEM================================')
    mode= input('\nAre you a Student or an Admin? \n[Student/ Admin]: ')
    if mode.lower()== 'student' or 'admin':
        stop_tier_2(mode)
        architecture_tier_2(mode.lower(), 'database')
    else:
        print('\nInvalid Mode')
        stop_tier_2()
        
        
def architecture_tier_2(mode, database):
    import random
    global clearance
    first= input('\nEnter your First Name: ').title()
    stop_tier_2(first)
    last= input('\nEnter your Last Name: ').title()
    stop_tier_2(last)
    email= input('\nEnter your Email: ')
    i_d= input('\nEnter your ID ("None" if you\'re not a member yet): ')
    stop_tier_2(i_d)
    user= DSC_database(first= first, last=last, email= email)
    
    if i_d.lower()== 'none':
        print('Looks like you need to sign up!\n')
        confirm= input('Do you want to sign up? \n[Yes/ No]:')
        if confirm.lower()== 'yes':
                i_d= user.sign_up(mode)
                ask= input('Do you wish to continue: ')
                if ask.lower()=='yes': return architecure_tier_1(); return 'okay then, bye'
        else:
            return 'okay then, bye'

    elif i_d in getattr(DSC_database, database).index and first+' '+last==\
    getattr(DSC_database, database).loc[i_d].Name:
        print(f'\nwelcome {user.Fullname()}\n')
        print(clearance)
        request= int(input('Enter the action you desire to perform: \n'))
        
        if request == 1:
            if mode.lower()== 'admin':
                mode= input('Enter the mode you wish to sign up: ')
                stop_tier_2(mode)
                first= input('\nFirst Name: ').title()
                stop_tier_2(first)
                last= input('\nLast Name: ').title()
                stop_tier_2(last)
                email= input('\nEmail: ')
                user= DSC_database(first= first, last=last, email= email)
                i_d= user.sign_up(mode)
            else:
                i_d= user.sign_up('student')
                first= input('\nFirst Name: ').title()
                stop_tier_2(first)
                last= input('\nLast Name: ').title()
                stop_tier_2(last)
                email= input('\nEmail: ')
                user= DSC_database(first= first, last=last, email= email)
            stop_tier_1()
                
                
        elif request == 2:
            name= input('Enter new name: ')
            email= input('Enter new Email: ')
            user.update_data(i_d, name, email)
            stop_tier_1()
            
        elif request == 3:
            i_d_needed= input('Enter the ID the person you wish to retrive their data')
            user.retrive_data(i_d_needed, mode)
            stop_tier_1()
            
        elif request == 4:
            if mode.lower()== 'admin':
                id_needed= input('Enter the ID you need to delete: ')
                user.delete_data(id_needed)
            else:
                print('Sorry, you\'re not cleared for this action' )
            stop_tier_1()
                
        elif request == 5:
            user.view_data(mode, i_d)
            stop_tier_1()
            
        else:
            print('Invalid request')
            stop_tier_1()
            
    else:
        print('Invalid Log in database')
        stop_tier_1()
```
